Program Python untuk mengonversi daftar karakter

```python
def convert(s):
 new = ""
 for x in s:
    new += x
 return new

# driver code
s = ['g', 'e', 'e', 'k', 's', 'f', 'o', 'r',
'g', 'e', 'e', 'k', 's']
print(convert(s))
```
